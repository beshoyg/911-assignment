```python
import pandas as pd
```

```python
df = pd.read_csv('911.csv')
```

```python
df.head()
```

```python
#1 How many calls for Vehicle accidents happened?

df[df['title']=='Traffic: VEHICLE ACCIDENT -'].count()
```

```python
#2 which town has the most 911 calls and what is the total number of calls for this town
df['twp'].value_counts()[:1]
```

```python
#3 what is/are the address/addresses for the 911 calls at the following date and time 2015-12-10 17:40:00
df[df['timeStamp'] == '2015-12-10 17:40:00']['addr']
```

```python
#4 how many calls at a longitude -75.5812935 and latitude 40.2978759
df[(df['lat'] == 40.2978759) & (df['lng']== -75.5812935)].count()
```

```python
# how many disabled vehicle accidents's calls took place for address located at Stump RD and Horsham RD
df[(df['addr'] == 'STUMP RD & HORSHAM RD') & (df['title'] == 'Traffic: DISABLED VEHICLE -')].count()
```

```python
#6 list all accidents with the title "Traffic: Vehicle Accident" that took place at Plymouth town
df[(df['title']=='Traffic: VEHICLE ACCIDENT -') & (df['twp'] == 'PLYMOUTH')]
```

```python
#7 list all accidents with a title "Traffic: VEHICLE ACCIDENT -" that took place at or after date and time "2016-08-23 16:17:00"
df[(df['title']=='Traffic: VEHICLE ACCIDENT -') & (df['timeStamp'] >= '2016-08-23 16:17:00')]
```
